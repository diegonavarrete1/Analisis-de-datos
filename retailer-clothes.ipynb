```python
avior= avior.dropna(subset= 'Review Rating')
avior= avior.dropna(subset= 'Age')
avior= avior.dropna(subset= 'Shipping Type')
print(avior)
avior.rename(columns={'Purchase Amount (USD)': 'Purchase USD'}, inplace=True)

sells_per_season = avior.groupby('Season')['Purchase USD'].sum()
print(sells_per_season / sells_per_season.sum())

item_sells = avior.groupby('Item Purchased')['Purchase USD'].sum()
item_prop = item_sells / item_sells.sum()
print(item_prop.sort_values(ascending=False))

sizes = avior.groupby('Size')['Purchase USD'].sum()
print((sizes / sizes.sum()).sort_values(ascending=False))

color = avior.groupby('Color')['Purchase USD'].sum()
print((color / color.sum()).sort_values(ascending=False))
bins= [i*10 for i in range(0,11)]
labels= [f'{i*10}-{(i+1)*10}' for i in range(0,10)]
avior['Rangos']= pd.cut(avior['Age'], bins= bins, labels= labels)
edad= avior.groupby('Rangos')['Purchase USD'].sum()
print(edad/edad.sum())
print((avior[avior['Rangos']== '60-70']['Item Purchased'].value_counts(ascending= False)))
print(avior['Age'].corr(avior['Review Rating']))
print(avior['Purchase USD'].corr(avior['Review Rating']))
print(print(avior['Purchase USD'].corr(avior['Previous Purchases'])))
gender= avior.groupby('Gender')['Purchase USD'].sum()
location= avior.groupby('Location')['Purchase USD'].sum()
print('gender proportion:',gender/ gender.sum())
print('location proportion:', location/ location.sum())
subs_ratio = avior['Subscriptionn Status'].value_c
```

```python
subs_ratio = avior['Subscription Status'].value_counts()
print(subs_ratio/ subs_ratio.sum())
subs_spend = avior.groupby('Subscription Status')['Purchase USD'].mean()
print(subs_spend)
```
