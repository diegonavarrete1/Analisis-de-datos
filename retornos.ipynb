```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import pandas as pd
import os
finance = pd.read_csv('/kaggle/input/datasets/belbino/global-financial-markets-dataset-2010-present/global_financial_markets_2000_Now.csv')
print(finance.head())
import pandas as pd
import os


finance = pd.read_csv('/kaggle/input/datasets/belbino/global-financial-markets-dataset-2010-present/global_financial_markets_2000_Now.csv')


pd.set_option('display.float_format', '{:,.4f}'.format)


# =========================
finance = finance[finance['volume'] != 0]

#

# Retorno
finance['returns'] = (finance['close'] - finance['open']) / finance['open']

# Comportamiento precio
finance['behaviour'] = finance['close'] / finance['open']

# Rango diario
finance['tails'] = finance['high'] / finance['low']

# "Profit" ponderado (no real, pero útil)
finance['profits'] = finance['returns'] * finance['volume']


# Conteo de activos
print("\nActivos más frecuentes:")
print(finance['asset_name'].value_counts().head(10))

# Tipos de activos
print("\nTipos de activos:")
print(finance['asset_type'].value_counts())

# Profit total por activo
prof = finance.groupby('asset_name')['profits'].sum().sort_values(ascending=False)
print("\nTop activos por 'profits':")
print(prof.head(10))

total_types = finance.groupby('asset_type')['profits'].sum()

print("\nProporción por tipo de activo:")
print(total_types / total_types.sum())

# 
proportions = finance.groupby(['asset_type','asset_name'])['profits'].sum()

proportions = proportions / proportions.groupby(level=0).transform(lambda x: x.abs().sum())

print("\nProporción por activo dentro de su tipo:")
print(proportions.sort_values(ascending=False).head(20))

finance['volatility']= finance.groupby('asset_name')['returns'].transform('std')
risk_return= finance.groupby('asset_name')['returns'].agg(['mean', 'std'])
print(risk_return.sort_values('mean', ascending= False))
risk_return['sharpe']= risk_return['mean']/ risk_return['std']
print(risk_return['sharpe'].sort_values(ascending= False))
print("\nPreview final:")
print(finance.head())
```
